```python
from jiwer import wer
import whisper

# Load the Whisper model
model = whisper.load_model("base")

# Define audio file paths and corresponding reference transcripts
audio_files = ["/content/Ali Sethi  Rung (Official Audio).mp3.wav.wav", "/content/Anuv Jain - ALAG AASMAAN (a song on the ukulele).mp3.wav.wav"]  # Replace with your audio file paths
references = ["Hello this is aayan", "...."]  # Replace with corresponding reference texts

# Initialize a dictionary to store WERs
wers = {}

# Calculate WER for each audio file
for audio_file, reference in zip(audio_files, references):
    result = model.transcribe(audio_file)
    prediction = result["text"]

    error = wer(reference, prediction)  # Note: Order is reference, prediction

    wers[audio_file] = error
    print(f"WER for {audio_file}: {error}")

# Find the audio file with the lowest WER (best accuracy)
best_audio = min(wers, key=wers.get)

print(f"\nAudio file with the best accuracy (lowest WER): {best_audio} (WER: {wers[best_audio]})")
```
